```python
# Mean of residuals ~ 0 (use the same design matrix that trained 'result')
import statsmodels.api as sm
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Design matrix used in training
X_train_design = X_sm.copy()  # already has constant
yhat_train = result.predict(X_train_design)
resid = Y_train - yhat_train

print(f"Residual mean: {float(resid.mean()):.6f}")
```

```python
# Homoscedasticity: Goldfeld–Quandt test using the full design matrix (must be 2D)
from statsmodels.stats.diagnostic import het_goldfeldquandt

gq = het_goldfeldquandt(resid, X_train_design, alternative="two-sided")
F_stat, p_val, alt = gq
print(f"Goldfeld–Quandt: F={F_stat:.4f}, p={p_val:.4f}, alternative='{alt}'")
print("No strong evidence of heteroscedasticity (homoscedasticity validated)." if p_val > 0.05
      else "Potential heteroscedasticity detected; consider robust SE (HC3).")
```

```python
# Normality: histogram and Q-Q plot
sns.histplot(resid, kde=True, stat="density")
plt.title("Residuals distribution (~bell-shaped expected)")
plt.xlabel("Residuals")
plt.show()

sm.qqplot(resid, line="45", fit=True)
plt.title("Q-Q Plot of Residuals (~on the line expected)")
plt.show()
```

```python
# Evaluation on the standardized test set used in the notebook
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Build test design matrix with constant; columns must match training
X_test_design = sm.add_constant(X_test[X_train.columns], has_constant="add")
y_pred_test = result.predict(X_test_design)

mae = mean_absolute_error(Y_test, y_pred_test)
mse = mean_squared_error(Y_test, y_pred_test)  # version-safe; returns MSE
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, y_pred_test)

n, p = X_test.shape
adj_r2 = 1 - (1 - r2) * (n - 1) / max(n - p - 1, 1)

print(f"Test MAE : {mae:.4f}")
print(f"Test RMSE: {rmse:.4f}")
print(f"Test R²  : {r2:.4f}")
print(f"Test AdjR²: {adj_r2:.4f}")

# Optional: save predictions
pd.DataFrame({"y_true": Y_test, "y_pred": y_pred_test}).to_csv("predictions_test.csv", index=False)
```
